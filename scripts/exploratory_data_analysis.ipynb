```python
data = pd.read_csv('../data/listings_preprocessed.csv')
```

```python
data.head(5)
```

```python
# visualize binary features
for c in data.columns:
    if len(data[c].value_counts()) == 2:
        print('feature name:', c)
        print(data[c].value_counts(), '\n')
```

```python
# visualize columns to be one-hot encoded
sns.set(style="darkgrid")

columns_to_binarize = []

for c in data.select_dtypes(include=[object]).columns:
    columns.append(c)
    if len(set(data[c])) > 50:
        plt.figure(figsize=(10,50), dpi=150)
        ax = sns.countplot(y=c, data=data)
    elif 10 < len(set(data[c])) < 50:
        plt.figure(figsize=(10,10))
        ax = sns.countplot(y=c, data=data)
    else:
        plt.figure(figsize=(10,5))
        ax = sns.countplot(x=c, data=data)
    plt.show()

print(columns_to_binarize)
```

```python
data[columns_to_binarize].head(3)
```

```python
one_hot_encoding = pd.get_dummies(data[columns_to_binarize])
one_hot_encoding.head(3)
```

```python
data = pd.concat([data, one_hot_encoding], axis=1)
data.head(3)
```

```python
data.drop(columns_to_binarize, axis=1, inplace=True)
```

```python
sorted(list(data.columns))
```

```python
y = data['price']
X = data.drop(['price'], axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X.shape
```

## Generalized Linear Models

```python
# Linear models may not well capture the data in high dimension.
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(normalize=True, n_jobs=-1)
scores = cross_val_score(reg, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
from sklearn.linear_model import Lasso
reg_l1 = Lasso(alpha=0.1)
scores = cross_val_score(reg_l1, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
from sklearn.linear_model import Ridge
reg_l2 = Ridge(alpha=0.1)
scores = cross_val_score(reg_l2, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
from sklearn.linear_model import ElasticNet
e_net = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=0, warm_start=True)
scores = cross_val_score(e_net, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

## Ensemble Models

```python
%%time
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, criterion='mae', max_features='sqrt', max_depth=None, n_jobs=-1)
scores = cross_val_score(rf, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
%%time
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(loss='ls', learning_rate=0.05, n_estimators=100, subsample=1.0, 
                               criterion='friedman_mse', min_samples_split=2, max_depth=20, 
                               validation_fraction=0.1, tol=0.0001)
scores = cross_val_score(gb, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
%%time
from sklearn.ensemble import BaggingRegressor
b = BaggingRegressor(base_estimator=None, n_estimators=100, max_samples=1.0, max_features=1.0, bootstrap=True, 
                     n_jobs=-1, random_state=42, verbose=1)
scores = cross_val_score(b, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

```python
%%time
from sklearn.ensemble import AdaBoostRegressor
ab = AdaBoostRegressor(base_estimator=None, n_estimators=10, learning_rate=0.01, loss='square', random_state=42)
scores = cross_val_score(ab, X, y, cv=5)
print('R^2 score:', np.average(scores[scores > 0]))
```

### LigthGBM

```python
# One of the great features of LightGBM is to visualize the feature importance ranking. Although the rankings 
# may not be fully reasonable, it provides some insights on the importance of certain features.
```

```python
import lightgbm as lgb
```

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
```

```python
params = {
    'boosting_type': 'dart',
    'objective': 'regression',
    'num_iterations': 200,
    'metric': {'l2', 'l1', 'mape'},
    'learning_rate': 0.1,
    'verbose': 1
}
```

```python
gbm = lgb.train(params,
                lgb_train,
                num_boost_round=200,
                valid_sets=lgb_eval,
                early_stopping_rounds=10)
```

```python
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE:', mean_squared_error(y_test, y_pred) ** 0.5)
```

```python
from sklearn.metrics import r2_score
print('R^2 of test set', r2_score(y_test, y_pred))
```

```python
%matplotlib inline

fig, ax = plt.subplots(1,1,figsize=(10,30), dpi=150)
lgb.plot_importance(gbm, ax=ax)
# fig.savefig('feature_importance.png')
```
